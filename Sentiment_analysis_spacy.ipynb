## 2. Remove special chracters and hyperlinks from text.

```python
#importing the relevent libraries

import re
import string
```

```python
# we are defining the function to find the special charater and hyperlinks from the sentences
def preprocess(sentence):
    pattern = r'[^a-zA-Z0-9\s]' # defining the pattern to eliminating the special character by keeping only the alpha(lower and upper). numerical values and spaces
    rem_spec = re.sub(pattern,"",sentence) #substituting pattern with spaces
    rem_hyper = re.sub(r'http\S+', '',rem_spec) # removing hyperlinks
    rem_br = re.sub(r'br', '',rem_hyper) # and also removing the <br>
    return rem_br
    
```

```python
df['review_cleaned'] = df['review'].apply(preprocess) #applying the function to the dataset
```

## Using Spacy Find the user names,City mentioned in the top 5000 records and store in list and add another column to list called Entity like if name is name of person then new column should contain "Person" if name is name of City then new column should be "City"

```python
# importing the relevent libraries
import spacy #importing the spacy for performing nlp tasks
nlp = spacy.load('en_core_web_sm') #model for the English language is en_core_web_sm.
```

Named Entity Recognition (NER) is the process of locating named entities in unstructured text and then classifying them into pre-defined categories, such as person names, organizations, locations, monetary values, percentages, time expressions, and so on.

```python
# we are defining the entity finder function 
def entity_finder(sentence):
    doc = nlp(sentence) #this method to convert any text into a processed Doc object and deduce attributes
    entity_doc = [[ent.text,ent.label_] for ent in doc.ents if ent.label_ == "PERSON" or ent.label_ == "GPE"]
    return entity_doc
            
```

```python
new = df['review_cleaned'][:5000].apply(entity_finder) #applying the entity_finder for our dataset
new
```

```python
# Making a list of entities
entity_list = []
for i in range(len(new)):
    for j in range(len(new[i])):
        entity_list.append(new[i][j])
        print(new[i][j])
```

```python
#Creating a dataframe consisting of Name and their respective type of entity
new_df = pd.DataFrame(entity_list,columns = ['Name','Entity'] )
new_df
```

```python
#Converting the dataframe to json file
def to_json(dataframe):
    json = dataframe.to_json() 
    print(json) 
```

```python
to_json(new_df)
```
